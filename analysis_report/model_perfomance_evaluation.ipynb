```python
import matplotlib.pyplot as plt
import numpy as np

# Performance values
models = ['Prophet', 'ANN', 'RNN', 'LSTM']
mae_values = [202.2529597226919, 2104.3777184300766, 872.3803176415139, 665.8794001708337]
mse_values = [70248.34280903936, 5650167.739974697, 787569.3854014337, 447751.8112231438]
rmse_values = [265.04403937655223, 2377.0081489079284, 887.4510608486722, 669.1425940882435]

# Plotting the MAE
plt.figure(figsize=(10, 3))
plt.bar(models, mae_values, color='skyblue')
plt.ylabel('Mean Absolute Error (MAE)')
plt.title('Model Performance Comparison - MAE')
plt.xlabel('Models')
plt.tight_layout()
plt.show()
```

```python
# Plotting the MSE
plt.figure(figsize=(10, 3))
plt.bar(models, mse_values, color='salmon')
plt.ylabel('Mean Squared Error (MSE)')
plt.title('Model Performance Comparison - MSE')
plt.xlabel('Models')
plt.tight_layout()
plt.show()
```

```python
# Plotting the RMSE
plt.figure(figsize=(10, 3))
plt.bar(models, rmse_values, color='lightgreen')
plt.ylabel('Root Mean Squared Error (RMSE)')
plt.title('Model Performance Comparison - RMSE')
plt.xlabel('Models')
plt.tight_layout()
plt.show()
```
